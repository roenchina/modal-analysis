```python

import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile


file_name = './EXP_srtp/_T353 glass 0.5/T353-p190-Glass-s0.5_gt.wav'
# file_name = './EXP_srtp/_T150 Wood s10/T150-p196_test.wav'

y, sr = librosa.load(file_name, sr=None, duration=1)
# sr, y = wavfile.read(file_name)
plt.figure(figsize=(8, 8))

plt.subplot(2,1,1)
plt.ylim((-500, 500))
frame = plt.gca()
frame.axes.get_xaxis().set_visible(False)
librosa.display.waveshow(y[0:22050], sr=sr)
plt.ylabel('Amplitude')

plt.subplot(2,1,2)
# option 1
Pxx, freqs, bins, im = plt.specgram(y, NFFT=1028, Fs=sr)
plt.axis((None, None, 0, 20000))

# option 2
# D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
# librosa.display.specshow(D, x_axis='s', y_axis='mel')

# plt.colorbar(format='%+2.0f dB')
plt.ylabel('Mel Scale')

plt.show()
```
